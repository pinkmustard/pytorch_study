```python
# 문장의 원-핫 벡터 만들기 (TF표현)
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns

corpus = ['Time flies like an arrow.',
          'Fruit flies like a banana']
one_hot_vectorrizer = CountVectorizer(binary=True)
one_hot = one_hot_vectorrizer.fit_transform(corpus).toarray()
vocab = one_hot_vectorrizer.get_feature_names_out()
sns.heatmap(one_hot, annot=True,
            cbar=False, xticklabels=vocab,
            yticklabels=['Sentence1', 'Sentence2'])
```

```python
# 문장의 원-핫 벡터 만들기 (TF-IDF표현)
from sklearn.feature_extraction.text import TfidfVectorizer
import seaborn as sns

tfidf_vectorizer = TfidfVectorizer()
tfidf = tfidf_vectorizer.fit_transform(corpus).toarray()
sns.heatmap(tfidf, annot=True,
            cbar=False, xticklabels=vocab,
            yticklabels=['Sentence1', 'Sentence2'])
```
